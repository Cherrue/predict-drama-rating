```python
%matplotlib inline
import mglearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn import svm
```

```python
data=pd.read_excel('cable_text.xlsx','predictSheet')
```

```python
attr_X=[]
temp=[]
for i in range(len(data)):
    temp=[data.ix[i,'day1'],data.ix[i,'day2'],data.ix[i,'day3'],data.ix[i,'day4'],data.ix[i,'day5'],data.ix[i,'day6'],data.ix[i,'day7']]
    attr_X.append(temp)
attr_X = np.array(attr_X)
attr_y = data['rating']
attr_y = np.array(attr_y)
```

```python
#We call our estimator instance clf, as it is a classifier.
clf = svm.SVR()
print('fit',clf.fit(attr_X, attr_y))
# given data to predict value
print('predict',clf.predict([[2., 2., 2., 2., 2., 2., 2.]]))
# evaluation score R squared
print('score',clf.score(attr_X,attr_y))
print('parameter',clf.get_params)
```
